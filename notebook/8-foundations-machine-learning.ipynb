```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

sys.path.insert(0, str(Path().resolve().parent))
from config import DATA_RAW, DATA_PROCESSED, FIGURES, RANDOM_SEED

sns.set_theme(style="whitegrid", palette="muted")
rng = np.random.default_rng(RANDOM_SEED)

from sklearn.datasets import load_iris, fetch_california_housing
from sklearn.model_selection import train_test_split

# ── Classification dataset: Iris ───────────────────────────────────────────
iris = load_iris(as_frame=True)
X_iris = iris.data
y_iris = iris.target   # 0=setosa, 1=versicolor, 2=virginica

# ── Regression dataset: California Housing ────────────────────────────────
housing = fetch_california_housing(as_frame=True)
X_housing = housing.data
y_housing = housing.target   # median house value (×$100,000)

print("=== Iris (classification) ===")
print(f"  Observations : {X_iris.shape[0]}")
print(f"  Features     : {X_iris.shape[1]}  ({list(X_iris.columns)})")
print(f"  Classes      : {list(iris.target_names)}")

print("\n=== California Housing (regression) ===")
print(f"  Observations : {X_housing.shape[0]:,}")
print(f"  Features     : {X_housing.shape[1]}  ({list(X_housing.columns)})")
print(f"  Target range : [{y_housing.min():.2f}, {y_housing.max():.2f}]  "
      f"(median house value ×$100k)")
```
